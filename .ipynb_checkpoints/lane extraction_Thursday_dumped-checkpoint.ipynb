````python
def roi(img, vertices):
    mask = np.zeros_like(img)
    cv2.fillPoly(mask, vertices, 255)
    masked = cv2.bitwise_and(img, mask)
    return masked
````

## color clustering

````python
def centroid_histogram(clt):
    # grab the number of different clusters and create a histogram
    # based on the number of pixels assigned to each cluster
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins = numLabels)
    # normalize the histogram, such that it sums to one
    hist = hist.astype("float")
    hist /= hist.sum()
    # return the histogram
    return hist
````

````python
def plot_colors(hist, centroids):
    # initialize the bar chart representing the relative frequency
    # of each of the colors
    bar = np.zeros((50, 300, 3), dtype = "uint8")
    startX = 0

    # loop over the percentage of each cluster and the color of
    # each cluster
    for (percent, color) in zip(hist, centroids):
        # plot the relative percentage of each cluster
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX
    # return the bar chart
    return bar
````

## counter most_common 
https://github.com/python/cpython/blob/3.6/Lib/heapq.py
```python
    def most_common(self, n=None):
        '''List the n most common elements and their counts from the most
        common to the least.  If n is None, then list all element counts.
        >>> Counter('abcdeabcdabcaba').most_common(3)
        [('a', 5), ('b', 4), ('c', 3)]
        '''
        # Emulate Bag.sortedByCount from Smalltalk
        if n is None:
            return sorted(self.iteritems(), key=_itemgetter(1), reverse=True)
        return _heapq.nlargest(n, self.iteritems(), key=_itemgetter(1))

```

````python
from collections import Counter
def count_color(width, height,image):
    manual_count={}
    for y in range(0, width):
        for x in range(0,height):
            rgb = (image[x,y,2],image[x,y,1],image[x,y,0])
            if rgb in manual_count:
                manual_count[rgb][0] +=1
                manual_count[rgb][1].append((x,y))
            else:
                manual_count[rgb] =[1,[(x,y)]]
    number_counter = Counter(manual_count).most_common(6)                
    return manual_count, number_counter
````

````python
def draw_mask_image(img, lines=None):
    nrows, ncolums = img.shape
    canvans = np.zeros((nrows, ncolums, 3), dtype = "uint8")
    canvans = cv2.rectangle(canvans, (int(0), 0), (int(nrows),ncolums),
                               np.array((0,0,0),dtype = "uint8").tolist(), -1)
    return cv2.cvtColor(canvans, cv2.COLOR_BGR2GRAY)
````

````python
HEIGHT = 270
WIDTH=480
ANGLE_THRESHOLD=30

def distance_line_line(l1,l2):
    """ distance calculate the distance of two line at y=y_max, actually it is the distance between two points
    actually it is not the perpenticular distance between two lines.
    distance = x2 - x1:
    # if x2 - x1 > 0: then line 2 is on the right side of line 1
    # if x2 - x1 < 0: then line 2 is on the left side of line 1
    """ 
    m1 = l1[0]
    m2 = l2[0]
    b1 = l1[1]
    b2 = l2[1]
    max_y = HEIGHT
    x1 = (max_y-b1) / m1 # NEED REVISED
    x2 = (max_y-b2) / m2 # max_yheight
    return x2-x1

def distance_point_line(point,l):
    """calculate the distance between point and l"""

# def process_lines_add_angle(lines):
#     """return slope, angle, [x1,y1,x2,y2]"""
#     new_lines = []
#     lines_angle_dict = {}
#     min_y = 0
#     max_y = HEIGHT
#     for idx,i in enumerate(lines):
#         for xyxy in i:
#             # These four lines:
#             # modified from http://stackoverflow.com/questions/21565994/method-to-return-the-equation-of-a-straight-line-given-two-points
#             # Used to calculate the definition of a line, given two sets of coords.
#             x_coords = (xyxy[0],xyxy[2])
#             y_coords = (xyxy[1],xyxy[3])
#             A = vstack([x_coords,ones(len(x_coords))]).T
#             m, b = lstsq(A, y_coords)[0]
            
#             # in degree0
#             angle = np.arctan(m)/np.pi*180
#             # Calculating our new, and improved, xs 
#             x1 = (min_y-b) / m
#             x2 = (max_y-b) / m
#             if x1<x2:
#                 lines_angle_dict[idx] = [m,angle,[int(x1), min_y, int(x2), max_y]]
#                 new_lines.append([int(x1), min_y, int(x2), max_y])
#             elif x1>=x2:
#                 lines_angle_dict[idx] = [m,angle,[int(x2), max_y, int(x1), min_y]]
#                 new_lines.append([int(x2), max_y, int(x1), min_y])               

#     return lines_angle_dict,new_lines
def process_lines(lines):
    """return slope, angle, [x1,y1,x2,y2]"""
    new_lines = []
    lines_dict = {}
    min_y = 0
    max_y = HEIGHT
    for idx,i in enumerate(lines):
        for xyxy in i:
            # These four lines:
            # modified from http://stackoverflow.com/questions/21565994/method-to-return-the-equation-of-a-straight-line-given-two-points
            # Used to calculate the definition of a line, given two sets of coords.
            x_coords = (xyxy[0],xyxy[2])
            y_coords = (xyxy[1],xyxy[3])
            A = vstack([x_coords,ones(len(x_coords))]).T
            m, b = lstsq(A, y_coords)[0]
            
            # in degree0
            #angle = np.arctan(m)/np.pi*180
            # Calculating our new, and improved, xs 
            x1 = (min_y-b) / m
            x2 = (max_y-b) / m
            if x1<x2:
                lines_dict[idx] = [m,b,[int(x1), min_y, int(x2), max_y]]
                new_lines.append([int(x1), min_y, int(x2), max_y])
            elif x1>=x2:
                lines_dict[idx] = [m,b,[int(x2), max_y, int(x1), min_y]]
                new_lines.append([int(x2), max_y, int(x1), min_y])               

    return lines_dict,new_lines


def whether_intersect(yellow_lane_l,line_l):
    """input:
    yellow_lane_l = [m,angle,[x1, y1, x2, y2]
    line_angle_l = [m,angle,[x1, y1, x2, y2]
    """
    angle_y = yellow_lane_l[1]
    m_y = yellow_lane_l[0]
    b_y = yellow_lane_l[2][1]-m_y*yellow_lane_l[2][0] # b = y-m*x
    
    angle_l = line_angle_l[1]
    m_l = line_angle_l[0]
    b_l = line_angle_l[2][1]-m_l*line_angle_l[2][0] #  b = y-m*x
    
    if abs(angle_y-angle_l)<5: #
        return False
    else:
        x_intersect = -(b_y-b_l)/(m_y-m_l)
        y_intersect = (m_y*b_l-m_l*b_y)/(m_y-m_l)
        print(x_intersect,y_intersect)
        if x_intersect<WIDTH and x_intersect>0:
            if y_intersect<HEIGHT and y_intersect>0:
                return True #
            else:
                return False 
        else: 
            return False
        
        
def lines_clustering(line_dict):
    final_lanes = {}

    for idx in line_dict:
        final_lanes_copy = final_lanes.copy()
        m = line_dict[idx][0]
        b = line_dict[idx][1]
        line = line_dict[idx][2]

        if len(final_lanes) == 0:
            final_lanes[m] = [ [m,b,line] ]
        else:
            found_copy = False

            for other_ms in final_lanes_copy:
                if not found_copy:
                    if abs(other_ms*1.2) > abs(m) > abs(other_ms*0.8):
                        if abs(final_lanes_copy[other_ms][0][1]*1.2) > abs(b) > abs(final_lanes_copy[other_ms][0][1]*0.8):
                            final_lanes[other_ms].append([m,b,line])
                            found_copy = True
                            break
                    else:
                        final_lanes[m] = [ [m,b,line] ]
                        
    return final_lanes


# def lines_clustering(yellow_lane_angle,lines_angle_dict):
#     """
#     1.  ANGLE_THRESHOLD (4530)
#     2. 
#     3. y_maxx
#     4. 
#     5.  final_lanes
#     final_lanesdictionary
#     final_lanes['yellow']:{num:[slope,angle,[x1,y1,x2,y2]]}}
#     final_lanes['left']:{num:[slope,angle,[x1,y1,x2,y2]],
#                      num:[slope,angle,[x1,y1,x2,y2]],
#                      ...}
#     final_lanes['right']:{num:[slope,angle,[x1,y1,x2,y2]],
#                      num:[slope,angle,[x1,y1,x2,y2]],
#                      ...}
#     """
#     final_lanes = {}
#     final_lanes['yellow']=yellow_lane_angle
#     candidate = []

#     # x yellow_intersect_x
#     yellow_intersect_x = 400
#     yellow_angle = yellow_lane_angle[0][1]
#     if yellow_angle<=0:
#         yellow_intersect_x = yellow_lane_angle[0][2][0]
#     elif yellow_angle>0:
#         yellow_intersect_x = yellow_lane_angle[0][2][2]

#     # ANGLE_THRESHOLD
#     # yellow_intersect_x
#     candidate_left = []
#     candidate_right = []
#     for idx in lines_angle_dict:
#         #final_lanes_copy = final_lanes.copy()
#         angle = lines_angle_dict[idx][1]
#         # ANGEL_THRESHOLD
#         if abs(angle-yellow_lane_angle[0][1])<ANGLE_THRESHOLD:# and not whether_intersect(yellow_lane_angle[0],lines_angle_dict[idx]):
#             if distance_line_line(yellow_lane_angle[0][2],lines_angle_dict[idx][2])<0:
#                 candidate_right.append(lines_angle_dict[idx])
#             elif distance_line_line(yellow_lane_angle[0][2],lines_angle_dict[idx][2])>=0:
#                 candidate_left.append(lines_angle_dict[idx])

#     final_lanes['left'] = candidate_left
#     final_lanes['right'] = candidate_right
# #         if len(final_lanes) == 0:
# #             final_lanes[m] = [ [m,b,line] ]
# #         else:
# #             found_copy = False

# #             for other_ms in final_lanes_copy:
# #                 if not found_copy:
# #                     if abs(other_ms*1.2) > abs(m) > abs(other_ms*0.8):
# #                         if abs(final_lanes_copy[other_ms][0][1]*1.2) > abs(b) > abs(final_lanes_copy[other_ms][0][1]*0.8):
# #                             final_lanes[other_ms].append([m,b,line])
# #                             found_copy = True
# #                             break
# #                     else:
# #                         final_lanes[m] = [ [m,b,line] ]
#     return final_lanes
````

````python
def display_lanes(rgb_image):
    original_image = rgb_image   
    #gray_img = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)
    processed_img = cv2.Canny(rgb_image, threshold1=200, threshold2=300)
    processed_img = cv2.GaussianBlur(processed_img, (3,3), 0 )    
    vertices = np.array([[0,200],[0,0],[800,0],[800,200]], np.int32)
    processed_img = roi(processed_img, [vertices])
#    lines = cv2.HoughLinesP(processed_img, 1, np.pi/180, 180, 20, 15)
    lines = cv2.HoughLinesP(processed_img, 1, np.pi/180, 120, 20, 35)
    
    try:
        l1, l2 = find_main_lanes_straight(original_image,lines,1)
        cv2.line(original_image, (l1[0], l1[1]), (l1[2], l1[3]), [0,255,0], 10)
        cv2.line(original_image, (l2[0], l2[1]), (l2[2], l2[3]), [0,255,0], 10)
        cv2.circle(original_image,(480,270),10,[255,0,0])
#        for xyxy in lines:
            
#         print('line 1 point1 = ',l1[0],l1[1])
#         print('line 2 point1 = ',l2[0],l2[1])
#         print('line 1 point2 = ',l1[2],l1[3])
#         print('line 2 point2 = ',l2[2],l2[3])
        #cv2.circle(original_image,(l2[2],l2[3]),10,[0,255,0])
    except Exception as e:
        print(str(e))
        pass
    try:
        for coords in lines:
            coords = coords[0]
            try:
                cv2.line(processed_img, (coords[0], coords[1]), (coords[2], coords[3]), [255,0,0], 3)
            except Exception as e:
                print(str(e))
    except Exception as e:
        print(str(e))
    
    return original_image, processed_img
````
